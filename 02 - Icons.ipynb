# Icons

This notebook demonstrates how to use vendor icons in Databricks notebooks using **Simple Icons** and other icon sources.

## Simple Icons

**Simple Icons** provides free SVG icons for popular brands. You can access them via CDN:
<br />
<br />
```
https://cdn.simpleicons.org/{ICON_NAME}/{HEX_COLOR}
```

**Resources:**
- [Simple Icons Website](https://simpleicons.org/) - Browse all available icons
- [Simple Icons GitHub](https://github.com/simple-icons/simple-icons) - Source repository

## Other Icon Sources

Other sources for brand or developer icons are listed here:

- [Devicon](https://devicon-website.vercel.app/)
- [Icones/Iconify](https://icones.js.org/)
