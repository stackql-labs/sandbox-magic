## Simple Icons

**Simple Icons** provides free SVG icons for popular brands. You can access them via CDN:

```
https://cdn.simpleicons.org/{ICON_NAME}/{HEX_COLOR}
```

**Resources:**
- [Simple Icons Website](https://simpleicons.org/) - Browse all available icons
- [Simple Icons GitHub](https://github.com/simple-icons/simple-icons) - Source repository
