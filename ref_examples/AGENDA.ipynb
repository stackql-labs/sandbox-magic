## Course Overview

This course provides a comprehensive, structured approach to migrating from **{SOURCE_PLATFORM}** to **Databricks**. The migration journey is organized into six modules that follow the recommended phased approach, reducing risk and ensuring successful outcomes.

## Module Structure

| Module | Focus | Key Outcomes |
|--------|-------|--------------|
| [00 - Foundations]($./00 - Foundations/0.1 - Why Migrate to Databricks) | Why migrate, maturity stages, architecture mapping | Clear understanding of drivers and target state |
| [01 - Discover]($./01 - Discover/1.1 - Engagement and Alignment) | Engagement, stakeholder alignment, discovery, profiling, planning | Documented scope, dependencies, and roadmap |
| [02 - Design]($./02 - Design/2.1 - Target Architecture Design) | Target architecture, platform setup, storage, security, interoperability patterns | Environment ready, governance in place |
| [03 - Execute]($./03 - Execute/3.1 - Interoperability and Coexistence) | Data ingestion, replication and interoperability patterns, schema creation, CDC | Data landed and flowing in Databricks |
| [04 - Activate]($./04 - Activate/4.1 - Testing and Data Validation) | Data validation, observability, cutover execution, activation | Verified correctness, performance, and sign-off |
| [05 - Enable]($./05 - Enable/5.1 - Pipeline and Orchestration Migration) | Pipeline migration, code conversion, DevOps, BI/ML integration | Workloads operationalized on Databricks |
| [06 - Closeout]($./06 - Closeout/6.1 - Observability and Cost Monitoring) | Monitoring, enablement, documentation, decommissioning | Migration complete, source retired |
